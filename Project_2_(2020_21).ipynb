# Classification Model

Step 1: Import the Libraries and its various class or Functions

Step 2: Read the data From the given link.

Step 3: Define the Labels and Features.

Step 4: Scale the Features (As Greater Values can affect it Smaller values So scale from -1 to 1)

Step 5: Split the data in train and test

Step 6: Train the Model.

Step 7: Get the accuracy .

```
#import various Libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```
#read data from csv file
df=pd.read_csv('https://raw.githubusercontent.com/chaitanyabaranwal/ParkinsonAnalysis/master/parkinsons.csv')
df.head()
```

```
#define the features and labels
feature=df.loc[:,df.columns!='status'].values[:,1:]
labels=df.loc[:,'status'].values
#count the labels=1 and labels=0
print(labels[labels==1].shape[0],labels[labels==0].shape[0])
```

```
#scale features
scaler=MinMaxScaler((-1,1))
x=scaler.fit_transform(feature)
y=labels
#split the dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=7)
#train model
model=XGBClassifier()
model.fit(x_train,y_train)
```

```
#calculate Accuracy
y_predict=model.predict(x_test)
print("Accuracy : ",accuracy_score(y_test,y_predict)*100)
```
